```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import pandas as pd
```

```python
#importing Data
data=pd.read_csv("iris.csv")
```

```python
l=data.columns.size-1
for x in range(l)[::-1]:
    for y in range(x+1,l)[::-1]:
         #get Data from table
        attributes=data.iloc[50:149,0:4].to_numpy()
        classes=data.iloc[50:149,4].to_numpy().flatten()
        #train classificator
        clf = LinearDiscriminantAnalysis()
        clf.fit(attributes,classes)
        #plot
        plt.figure()
        plt.title('%s & %s' %(data.columns[x],data.columns[y]))
        plt.xlabel(data.columns[x])
        plt.ylabel(data.columns[y])
        W=0
        for i in range(50,100):
            C=clf.predict(data.iloc[i,0:4].to_numpy().reshape(1,4))[0]
            if C=='Versicolor':
                plt.plot(data.iloc[i,x],data.iloc[i,y],'bo',label='Versicolor')
            else:
                plt.plot(data.iloc[i,x],data.iloc[i,y],'go',label='wrongly classificated Versicolor')
                W=W+1
        for i in range(100,150):
            C=clf.predict(data.iloc[i,0:4].to_numpy().reshape(1,4))[0]
            if C=='Virginica':
                plt.plot(data.iloc[i,x],data.iloc[i,y],'g^',label='Virginica')
            else:
                plt.plot(data.iloc[i,x],data.iloc[i,y],'b^',label='wrongly classificated Virginica')
                W=W+1
plt.show()
```

```python
print('wrongly classificated=%d  (%d%%)'%(W,W))
```

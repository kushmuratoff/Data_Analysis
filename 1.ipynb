```python
#on first step we import the libraries we are going to work with;
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("iris.csv")
```

```python
#iris correlation for the whole table
print(data.corr())
```

```python
# we can notice their is a low correlation between sepal length and sepal width,
# meaning we can conclude they are statistically independent.
# we can also notice a height correlation between petal length and petal width,
# meaning we can conclude they are statistically correlated.
```

```python
# correlation for Setosa class
print(data.loc[0:49].corr())
```

```python
# it is noticeable that sepal length and sepal width are highly correlated,
# but not for petal length and width;
# which do not match with previous correlation table
# in general we should Expect in Setosa iris the petal width and length to be random,
# and for sepal length and width to correlated
```

```python
print(data.loc[50:99].corr())
```

```python
print(data.loc[100:149].corr())
```

```python
l=data.columns.size-1
for x in range(l)[::-1]:
    for y in range(x+1,l)[::-1]:
        plt.figure()
        plt.title('%s & %s' %(data.columns[x],data.columns[y]))
        plt.xlabel(data.columns[x])
        plt.ylabel(data.columns[y])
        plt.plot(data.iloc[0:49,x],data.iloc[0:49,y],'bo',label='Setosa')
        plt.plot(data.iloc[50:99,x],data.iloc[50:99,y],'ro',label='Versicolor')
        plt.plot(data.iloc[100:149,x],data.iloc[100:149,y],'co',label='Virginica')
        plt.legend()
plt.show()
```
